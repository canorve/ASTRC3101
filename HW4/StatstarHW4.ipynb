```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statstar import StatStar
```

#A

```python
model = pd.read_csv('starmodl_py.dat',sep='\s*')
```

```python
model
```

```python
model['logT'] = np.log10(model['T'])
model['logrho'] = np.log10(model['rho'])
model['logkap'] = np.log10(model['kap'])
```

```python
model
```

```python
model['kap_adjusted'] = 1.0*model['logrho'] + 24
```

```python
unmodified = model.plot(x='kap_adjusted',y='logkap', title=r'Unmodified log $\kappa$ vs. log $\rho$', style='k--')
model.plot(x='logrho',y='logkap')
unmodified.set_xlabel(r'log $\rho$')
unmodified.set_ylabel(r'log $\kappa$')
```

```python
model['kap_adjusted2'] = 0.5*model['logrho'] + 17
```

```python
modified = model.plot(x='kap_adjusted2',y='logkap', title=r'Modified log $\kappa$ vs. log $\rho$', style='k--')
model.plot(x='logrho',y='logkap')
modified.set_xlabel(r'log $\rho$')
modified.set_ylabel(r'log $\kappa$')
```

#B

```python
model['zones'] = model['zone'].replace('c', 1).replace('r', 0)
```

```python
zoneplot = model.plot(x='r', y='zones', title='Convective zone')
zoneplot.set_xlabel('Radius (cm)')
```

#C

```python
model['dT'] = model['T'].diff()
```

Tdiff = model.plot(x='r', y='dT', title='Temperature change vs Radius')
Tdiff.set_xlabel('Radius (cm)')

```python
model['dT2'] = model['dT'].diff()
Tdiff2 = model.plot(x='r', y='dT2', title='Temperature change vs Radius')
Tdiff2.set_xlabel('Radius (cm)')
```

```python
model['dlPdlTdiff'] = model['dlPdlT'].diff()
dlPdlTdiff = model.plot(x='r', y='dlPdlTdiff', title='Convectivity')
dlPdlTdiff.set_xlabel('Radius (cm)')
```

The figure shows convective zones far beyond where our model stops showing convectivity. In that sense, that figure may be more detailed than our simulation's categorization.

The outer envelope becomes convective for less massive stars, while the inner envelope becomes convective for more massive stars. This is due to the more massive stars burning through their MS lifetimes quicker. The Helium takes over more and more of the star from the core.
