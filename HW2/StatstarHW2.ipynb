## D

```python
modelA = pd.read_csv('starmodelA.dat', sep='\s*')
modelC = pd.read_csv('starmodelC.dat', sep='\s*')
```

```python
modelA
```

```python
modelC
```

```python
RvsRhoA = modelA.plot(x='r',y='rho', title='Density vs. Radius')

RvsRhoC = modelC.plot(x='r',y='rho', title='Density vs. Radius')
RvsRhoC.set_xlabel('Radius (cm)')
RvsRhoC.set_ylabel('Density (g/cm**3)')
```

```python
RvsTA = modelA.plot(x='r',y='T', title='Temperature vs. Radius')

RvsTC = modelC.plot(x='r',y='T', title='Temperature vs. Radius')
RvsTC.set_xlabel('Radius (cm)')
RvsTC.set_ylabel('Temperature (K)')
```

```python
RvsLA = modelA.plot(x='r',y='L_r', title='Luminosity vs. Radius')

RvsLC = modelC.plot(x='r',y='L_r', title='Luminosity vs. Radius')
RvsLC.set_xlabel('Radius (cm)')
RvsLC.set_ylabel('Luminosity (erg s**-1 )')
```

## E

```python
modelA['energyT'] = modelA['eps'] / (modelA['rho'] * (modelA['T']**4.5))
modelC['energyT'] = modelC['eps'] / (modelC['rho'] * (modelC['T']**4.5))
```

```python
epsRhoTA = modelA.plot(x='r',y='energyT', title='eps/rhoT vs. Radius')

epsRhoTC = modelC.plot(x='r',y='energyT', title='eps/rhoT vs. Radius')
epsRhoTC.set_xlabel('Radius (cm)')
epsRhoTC.set_ylabel('eps/rhoT**n')
```

## F

```python
modelA['logrho'] = np.log10(modelA['rho'])
modelA['logT'] = np.log10(modelA['T'])

modelC['logrho'] = np.log10(modelC['rho'])
modelC['logT'] = np.log10(modelC['T'])
```

```python
logRhologTA = modelA.plot(x='logT',y='logrho', title='Log Density vs. Log Temperature')

logRhologTC = modelC.plot(x='logT',y='logrho', title='Log Density vs. Log Temperature')
logRhologTC.set_xlabel('Log[T]')
logRhologTC.set_ylabel('Log[Rho]')
```
